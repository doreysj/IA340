````python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
````

````python
import matplotlib.pyplot as plt

# Assuming 'fips' is the unique identifier for each state
df.sort_values(by=['fips', 'year'], inplace=True)

# Calculate the year over year growth for each state's income
df['income_growth'] = df.groupby('fips')['income'].pct_change()

# Plot the income growth for each state over the years
for name, group in df.groupby('fips'):
    plt.plot(group['year'], group['income_growth'], label=name)

plt.xlabel('Year')
plt.ylabel('Income Growth Rate')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2)
plt.title('Year-over-Year Income Growth Rate for Each State')
plt.show()
````

````python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is the dataframe you mentioned
df['year'] = pd.to_datetime(df['year'], format='%Y')
df.set_index('year', inplace=True)
df.sort_index(inplace=True)

growth_rates = df.groupby('name')['income'].pct_change()*100
df['growth_rate'] = growth_rates

# Plotting
fig, ax = plt.subplots(figsize=(15, 7))
for key, grp in df.groupby(['name']):
    ax = grp.plot(ax=ax, kind='line', y='growth_rate', label=key)

plt.legend(loc='best')
plt.title('Year over year income growth rate per state')
plt.ylabel('Growth Rate (%)')
plt.xlabel('Year')
plt.show()
'''
This code will calculate the year-over-year income growth rate for each state and each year in the given dataframe. It then plots these growth rates on a line chart, with each state represented by a different line. The x-axis represents the year, and the y-axis represents the growth rate in percentage. The legend shows which line corresponds to which state.
````
